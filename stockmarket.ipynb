```python
from pandas_datareader import data
from datetime import datetime, timedelta, date
from bokeh.plotting import figure, show, output_file

ticker = input("Insert ticker: ")
start_date = datetime(2016, 3, 1)
end_date = datetime(2019, 9, 11)
aapl_data = data.DataReader(ticker, "yahoo", start_date, end_date)

graph = figure(x_axis_type="datetime", width=1000, height=300, 
               title=f"Candlestick chart for {ticker} from {start_date.date()} to {end_date.date()}", 
               x_axis_label="Date", y_axis_label="Price", sizing_mode="scale_width")
graph.grid.grid_line_alpha = 0.3

#lines
graph.segment(x0=aapl_data.index, x1=aapl_data.index, y0=aapl_data.Low, y1=aapl_data.High,
              line_color="black", line_width=1)
#positive
graph.quad(top=aapl_data.Close[aapl_data.Close >= aapl_data.Open], 
           bottom=aapl_data.Open[aapl_data.Close >= aapl_data.Open],
           left=(aapl_data.index[aapl_data.Close >= aapl_data.Open] - timedelta(hours=6)),
           right=(aapl_data.index[aapl_data.Close >= aapl_data.Open] + timedelta(hours=6)),
           fill_color="white", line_color="black")
#negative
graph.quad(top=aapl_data.Close[aapl_data.Close < aapl_data.Open], 
           bottom=aapl_data.Open[aapl_data.Close < aapl_data.Open],
           left=(aapl_data.index[aapl_data.Close < aapl_data.Open] - timedelta(hours=6)),
           right=(aapl_data.index[aapl_data.Close < aapl_data.Open] + timedelta(hours=6)),
           fill_color="black", line_color="black")
output_file("candlestick.html")
show(graph)
```
